A quick notebook that lets us explore the evaluation details for models

```python
import numpy as np
import sklearn as sk
import matplotlib as plt
import csv
import pandas as pd
```

```python
# varibales set for paths
path_bl = "baseline"
path_cl = "centreline"
path_cls = "centreline-speed"
path_e = "explorer"
path_w = "weighted"
csv_file = "0-iteration.csv"
```

```python
df_baseline = pd.read_csv("{}/{}".format(path_bl,csv_file))
```

```python
# read in csv files
df_cl_a = pd.read_csv("{}/a/{}".format(path_cl,csv_file))
df_cl_b = pd.read_csv("{}/b/{}".format(path_cl,csv_file))
df_cls_a = pd.read_csv("{}/a/{}".format(path_cls,csv_file))
df_cls_b = pd.read_csv("{}/b/{}".format(path_cls,csv_file))
df_e = pd.read_csv("{}/{}".format(path_e,csv_file))
df_w_a = pd.read_csv("{}/a/{}".format(path_w,csv_file))
df_w_b = pd.read_csv("{}/b/{}".format(path_w,csv_file))
```

```python
#check csv loaded
df_e.loc[df_e['episode_status']=="off_track"]
```

```python
print(len(df_e.loc[df_e['episode_status']=="off_track"]))
```

```python
# function to check how many times agent left track
def count_off_track(df,name):
    counter  = len(df.loc[df['episode_status']=="off_track"])
    print("{} left track {} times".format(name , counter))
    
```

```python
# show all off track for all agents evaluated
count_off_track(df_cl_a,"Centre Line A")
count_off_track(df_cl_b,"Centre Line B")
count_off_track(df_cls_a,"Centre Line Speed A")
count_off_track(df_cls_b,"Centre Line Speed B")
count_off_track(df_e,"Explorer")
count_off_track(df_w_a,"Weighted A")
count_off_track(df_w_b,"Weighted B")
```

```python
# total time for all agents evaluated
def total_time(df,name):
    total_time = df['tstamp'].iloc[-1]
    print("{} total time: {}".format(name, total_time))
```

```python
total_time(df_cl_b,"Centre Line B")
total_time(df_cls_a,"Centre Line Speed A")
total_time(df_cls_b,"Centre Line Speed B")
total_time(df_e,"Explorer")
total_time(df_w_a,"Weighted A")
total_time(df_w_b,"Weighted B")
```

```python
# total sets for all agents evaluated 
def total_steps(df,name):
    total_time = df['steps'].iloc[-1]
    print("{} total steps: {}".format(name, total_time))
```

```python
total_steps(df_cl_b,"Centre Line B")
total_steps(df_cls_a,"Centre Line Speed A")
total_steps(df_cls_b,"Centre Line Speed B")
total_steps(df_e,"Explorer")
total_steps(df_w_a,"Weighted A")
total_steps(df_w_b,"Weighted B")
```
